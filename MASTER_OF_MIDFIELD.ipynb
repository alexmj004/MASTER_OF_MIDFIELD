```python
# Generar gráfica comparativa de pases completados y % de acierto en los pases.
pedri_passes_completes = pedri_stats[['Statistic','Per 90']].query("Statistic == 'Passes Completed'")
pedri_passes_acierto = pedri_stats[['Statistic','Per 90']].query("Statistic == 'Pass Completion %'")
vitinha_passes_completes = vitinha_stats[['Statistic','Per 90']].query("Statistic == 'Passes Completed'")
vitinha_passes_acierto = vitinha_stats[['Statistic','Per 90']].query("Statistic == 'Pass Completion %'")

# Extraer y convertir a float explícitamente
pedri_passes_completes_val = float(pedri_passes_completes['Per 90'].values[0])
pedri_passes_acierto_val = float(pedri_passes_acierto['Per 90'].values[0])
vitinha_passes_completes_val = float(vitinha_passes_completes['Per 90'].values[0])
vitinha_passes_acierto_val = float(vitinha_passes_acierto['Per 90'].values[0])

labels = ['Passes Completed/90', 'Pass Completion %']
x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
fig.patch.set_alpha(0)  # Fondo transparente

# Barras con transparencia (alpha=0.7)
ax.bar(x - width/2, [pedri_passes_completes_val, pedri_passes_acierto_val], width,
       label='Pedri', color='#d62728', edgecolor='black', alpha=0.7)
ax.bar(x + width/2, [vitinha_passes_completes_val, vitinha_passes_acierto_val], width,
       label='Vitinha', color='#1f77b4', edgecolor='black', alpha=0.7)

# Personalización mejorada
ax.set_ylabel('Statistics', fontsize=12, labelpad=10)
ax.set_xticks(x)
ax.set_xticklabels(labels, fontsize=12)
ax.legend(loc='upper left', fontsize=12, framealpha=0.9)

# Añadir valores con sombra para mejor legibilidad
for i, (pedri_val, vitinha_val) in enumerate(zip(
    [pedri_passes_completes_val, pedri_passes_acierto_val],
    [vitinha_passes_completes_val, vitinha_passes_acierto_val]
)):
    # Texto para Pedri con sombra
    ax.text(x[i] - width/2, pedri_val + 0.5,
            f'{pedri_val:.1f}%' if i == 1 else f'{pedri_val:.1f}',
            ha='center', va='bottom', fontweight='bold',
            color='black', fontsize=10,
            bbox=dict(facecolor='white', alpha=0.7, edgecolor='none', boxstyle='round,pad=0.2'))

    # Texto para Vitinha con sombra
    ax.text(x[i] + width/2, vitinha_val + 0.5,
            f'{vitinha_val:.1f}%' if i == 1 else f'{vitinha_val:.1f}',
            ha='center', va='bottom', fontweight='bold',
            color='black', fontsize=10,
            bbox=dict(facecolor='white', alpha=0.7, edgecolor='none', boxstyle='round,pad=0.2'))

# Grid y estilo
ax.grid(axis='y', linestyle='--', alpha=0.4)
ax.set_axisbelow(True)  # Grid detrás de las barras

# Ajustar márgenes
plt.subplots_adjust(top=0.9, bottom=0.1)
plt.tight_layout()
plt.show()
```

```python
# Comparativa acciones defensivas de ambos jugadores.
tackles_won_pedri = pedri_stats[['Statistic','Per 90']].query("Statistic == 'Tackles Won'")
tackles_pedri = pedri_stats[['Statistic','Per 90']].query("Statistic == 'Tackles'")
tackles_attack_pedri = pedri_stats[['Statistic','Per 90']].query("Statistic == 'Tackles (Att 3rd)'")
tackles_mid_pedri = pedri_stats[['Statistic','Per 90']].query("Statistic == 'Tackles (Mid 3rd)'")
interceptions_pedri = pedri_stats[['Statistic','Per 90']].query("Statistic == 'Interceptions'")
passes_block_pedri = pedri_stats[['Statistic','Per 90']].query("Statistic == 'Blocks'")


# Datos
stats = [
    'Tackles Won',
    'Tackles',
    'Tackles (Att 3rd)',
    'Tackles (Mid 3rd)',
    'Interceptions',
    'Blocks'
]

values = [
    float(tackles_won_pedri['Per 90'].values[0]),
    float(tackles_pedri['Per 90'].values[0]),
    float(tackles_attack_pedri['Per 90'].values[0]),
    float(tackles_mid_pedri['Per 90'].values[0]),
    float(interceptions_pedri['Per 90'].values[0]),
    float(passes_block_pedri['Per 90'].values[0])
]

# Crear figura
fig, ax = plt.subplots(figsize=(10, 6), facecolor='white')

# Barras horizontales
bars = ax.barh(stats, values, height=0.6, color='#a50044', alpha=0.9, edgecolor='black', linewidth=0.7)

# Ajustar límites del eje X para espacio extra
ax.set_xlim(0, max(values) * 1.15)  # 15% más de espacio para las etiquetas

# Texto en barras
for bar in bars:
    width = bar.get_width()
    # Posición: ancho de barra + pequeño offset (3% del valor máximo)
    text_x = width + (max(values) * 0.03)
    ax.text(text_x,
            bar.get_y() + bar.get_height()/2,
            f'{width:.2f}',
            va='center',
            ha='left',
            fontsize=11,
            fontweight='bold',
            color='#333333')  # Color oscuro para mejor contraste

# Estilo de ejes
ax.set_xlabel('Actions per 90 min', fontsize=12, fontweight='bold', labelpad=10)
ax.set_ylabel('Statitics', fontsize=12, fontweight='bold', labelpad=10)
ax.tick_params(axis='both', which='major', labelsize=11)

# Remover bordes
for spine in ['top', 'right']:
    ax.spines[spine].set_visible(False)

# Grid horizontal sutil
ax.grid(axis='y', linestyle=':', alpha=0.4)

plt.tight_layout()
plt.show()
```

```python
# Comparativa acciones defensivas de Vitinha
tackles_won_vitinha = vitinha_stats[['Statistic','Per 90']].query("Statistic == 'Tackles Won'")
tackles_vitinha = vitinha_stats[['Statistic','Per 90']].query("Statistic == 'Tackles'")
tackles_attack_vitinha = vitinha_stats[['Statistic','Per 90']].query("Statistic == 'Tackles (Att 3rd)'")
tackles_mid_vitinha = vitinha_stats[['Statistic','Per 90']].query("Statistic == 'Tackles (Mid 3rd)'")
interceptions_vitinha = vitinha_stats[['Statistic','Per 90']].query("Statistic == 'Interceptions'")
passes_block_vitinha = vitinha_stats[['Statistic','Per 90']].query("Statistic == 'Blocks'")

# Datos
stats = [
    'Tackles Won',
    'Tackles',
    'Tackles (Att 3rd)',
    'Tackles (Mid 3rd)',
    'Interceptions',
    'Blocks'
]

values = [
    float(tackles_won_vitinha['Per 90'].values[0]),
    float(tackles_vitinha['Per 90'].values[0]),
    float(tackles_attack_vitinha['Per 90'].values[0]),
    float(tackles_mid_vitinha['Per 90'].values[0]),
    float(interceptions_vitinha['Per 90'].values[0]),
    float(passes_block_vitinha['Per 90'].values[0])
]

# Crear figura
fig, ax = plt.subplots(figsize=(10, 6), facecolor='white')

# Barras horizontales
bars = ax.barh(stats, values, height=0.6, color='#004d98', alpha=0.9, edgecolor='black', linewidth=0.7)

# Invertir el eje X para que vaya de derecha a izquierda
ax.invert_xaxis()

# Ajustar límites del eje X (ahora invertidos)
ax.set_xlim(max(values) * 1.15, 0)  # Máximo valor con 15% más de espacio

# Texto en barras (ahora a la izquierda)
for bar in bars:
    width = bar.get_width()
    # Posición: ancho de barra - pequeño offset (3% del valor máximo)
    text_x = width - (max(values) * -0.03)
    ax.text(text_x,
            bar.get_y() + bar.get_height()/2,
            f'{width:.2f}',
            va='center',
            ha='right',  # Alineado a la derecha
            fontsize=11,
            fontweight='bold',
            color='#333333')

# Estilo de ejes
ax.set_xlabel('Actions per 90 min', fontsize=12, fontweight='bold', labelpad=10)
ax.set_ylabel('Statistics', fontsize=12, fontweight='bold', labelpad=10)
ax.tick_params(axis='both', which='major', labelsize=11)

# Mover etiquetas del eje Y al lado derecho
ax.yaxis.set_label_position("right")
ax.yaxis.tick_right()

# Remover bordes
for spine in ['top', 'left']:  # Quitamos left en lugar de right
    ax.spines[spine].set_visible(False)

# Grid horizontal sutil
ax.grid(axis='y', linestyle=':', alpha=0.4)

plt.tight_layout()
plt.show()
```

```python
# Gráfico radar para comparar stats de creación de juego de Pedri y Vitinha.
pedri_progressive_passes = pedri_stats[['Statistic','Per 90']].query("Statistic == 'Progressive Passes'")
pedri_carries = pedri_stats[['Statistic','Per 90']].query("Statistic == 'Carries'")
pedri_sca = pedri_stats[['Statistic','Per 90']].query("Statistic == 'Shot-Creating Actions'")

vitinha_progressive_passes = vitinha_stats[['Statistic','Per 90']].query("Statistic == 'Progressive Passes'")
vitinha_carries = vitinha_stats[['Statistic','Per 90']].query("Statistic == 'Carries'")
vitinha_sca = vitinha_stats[['Statistic','Per 90']].query("Statistic == 'Shot-Creating Actions'")


# Extracción de datos para Pedri
pedri_pp = float(pedri_progressive_passes['Per 90'].values[0])
pedri_car = float(pedri_carries['Per 90'].values[0])
pedri_sca_val = float(pedri_sca['Per 90'].values[0])

# Extracción de datos para Vitinha
vitinha_pp = float(vitinha_progressive_passes['Per 90'].values[0])
vitinha_car = float(vitinha_carries['Per 90'].values[0])
vitinha_sca_val = float(vitinha_sca['Per 90'].values[0])

# Configuración de datos
stats = {
    'Progresive Passes': [pedri_pp, vitinha_pp],
    'Carriers': [pedri_car, vitinha_car],
    'Shot-Creating Actions': [pedri_sca_val, vitinha_sca_val]
}

categories = list(stats.keys())
values_pedri = [stats[k][0] for k in categories]
values_vitinha = [stats[k][1] for k in categories]

# Configuración del radar
angles = np.linspace(0, 2*np.pi, len(categories), endpoint=False).tolist()
angles += angles[:1]  # Cerrar el círculo

values_pedri += values_pedri[:1]
values_vitinha += values_vitinha[:1]

# Creación del gráfico
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw={'polar': True})

# Líneas de referencia
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories, fontsize=14)
ax.set_yticklabels([])
ax.set_theta_offset(np.pi/2)
ax.set_theta_direction(-1)

# Gráfico de Pedri (azul)
ax.plot(angles, values_pedri, 'b-', linewidth=3, label='Pedri')
ax.fill(angles, values_pedri, 'b', alpha=0.25)

# Gráfico de Vitinha (rojo)
ax.plot(angles, values_vitinha, 'r-', linewidth=3, label='Vitinha')
ax.fill(angles, values_vitinha, 'r', alpha=0.25)

# Ajustes estéticos
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1), fontsize=12)

# Añadir valores numéricos
for i, (angle, val_p, val_v) in enumerate(zip(angles[:-1], values_pedri[:-1], values_vitinha[:-1])):
    ax.text(angle, val_p*1.05, f'{val_p:.1f}', ha='center', va='center', color='blue', fontsize=10)
    ax.text(angle, val_v*1.05, f'{val_v:.1f}', ha='center', va='center', color='red', fontsize=10)

plt.tight_layout()
plt.show()
```
